```python
# import Libraries
import pickle
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg16 import preprocess_input
```

# Load Model

```python
# load model
model = load_model('/content/drive/MyDrive/model.h5')

with open('class.txt', 'r') as file:
    class_names = [line.strip() for line in file]
```

# Inferencing

## Test prediction

### Prediksi Jalan Rusak

kita akan memasukkan gambar Plothole atau jalan yang rusak

```python
# Memuat gambar dan mengubah ukurannya sesuai dengan input model

img_path = '/content/JalanRusak.jfif'
img = image.load_img(img_path, target_size=(150, 150))  # Ukuran model kita 150 x 150
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array = preprocess_input(img_array)

plt.imshow(img)
plt.show()
```

Hasil tes model pada gambar jalan rusak  berhasil memprediksi bahwa gambar tersebut merupakan Plothole

```python
# Make prediction and get class name
predictions = model.predict(img_array)
predicted_class_index = np.argmax(predictions, axis=-1)[0]
predicted_class_name = class_names[predicted_class_index]

# Print the predicted class name
print("Prediksi kelas:", predicted_class_name)
```

### Prediksi Jalan Mulus
